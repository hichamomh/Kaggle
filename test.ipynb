```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
distance =  []
distance_theory = []
for i in range(20):
    # serie positif
    random_series = np.random.rand(10)
    normalized_series = random_series/np.sum(random_series)
    serie_2 = [(normalized_series[i] - normalized_series[j])**2 for i in range(len(normalized_series)) for j in range(len(normalized_series))]
    dist = np.sum(serie_2)
    distance.append(dist)
    
    dist_theory = np.sum([normalized_series[i]-normalized_series[i]**2 for i in range(len(normalized_series))])
    distance_theory.append(-dist_theory)

plt.plot(distance_theory,distance)
plt.xlabel('theoretical distance')
plt.ylabel('distance')
plt.legend()
 
```

```python
def f(X):
    if np.sum(X)<=0.001:
        return np.zeros(len(X))
    return np.mean(X)+np.random.choice([-1,1],size=len(X))*f(np.abs(X-np.mean(X)))

# gnrate sample from normal distribution 5 1
X = np.random.normal(5,1,800)
_  = plt.hist(X)
plt.show()
new_X = f(X)
_ = plt.hist(new_X)
```

```python
def g(X):
    moyennes = []
    proba = []
    for i in range(300):
        moyennes.append(np.mean(X))
        prob1 = np.sum(X>=np.mean(X))/len(X)
        prob2 = np.sum(X<np.mean(X))/len(X)
        proba.append([prob1,prob2])
        X = np.abs(X-np.mean(X))
    return moyennes,proba

# random exponential distribution
X =  np.random.exponential(1,1000)
_ = plt.hist(X,bins=100)
plt.show()
moyennes,proba = g(X)
plt.plot(moyennes)
```

```python
new_X = []
for i in range(800):
    s = moyennes[0]
    for j in range(1,len(moyennes)):
        s += np.random.choice([1,-1],p=proba[j])*moyennes[j]
    if s>0:
        new_X.append(s)
_ = plt.hist(new_X,bins=100)
```
